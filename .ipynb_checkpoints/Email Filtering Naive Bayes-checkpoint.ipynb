```python
clean = df.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis = 1) # Drop last 3 columns
clean = df[['v1', 'v2']].rename(columns = {'v1': 'Category', 'v2': 'Message'}) # Rename to Category and Message
clean['Message'] = clean['Message'].str.lower() # Convert all letters to lowercase
clean['Message'] = clean['Message'].str.replace('[^a-z ]', '', regex = True) # Convert to only alpha characters
```

<h3>3. Split the Data</h3>
<p>Split the data into two random samples, one for training the model and the other for testing the model.  Create DataFrames named train_data and test_data.  The train_data DataFrame should contain 75% of the data and the test_data DataFrame the remaining 25%.<p>  


```python
train_data = clean.sample(frac = 0.75, random_state = 42)
test_data = clean.drop(train_data.index)
```

```python
common_words = ["the", "and", "or", "but", "if", "of", "a", "an", "in", "on", "at", "to", "for", "from", "with", "that", "this"] # Words that can't determine whether the email is spam or not as they are common words

def count_words(data, category):
    word_dict = {}
    # Filter data by category
    category_data = data[data['Category'] == category]
    for msg in category_data['Message']:
        # Split into a list of words
        words = msg.split()
        # Iterate over each word in the 'words' list for the current message
        for word in words:
            if word in common_words:
                continue
            if word in word_dict:
                word_dict[word] += 1
            else:
                word_dict[word] = 1
    
    sorted_word_dict = dict(sorted(word_dict.items(), key = lambda item: item[1], reverse = True))
    return sorted_word_dict

spam_dict = count_words(train_data, 'spam')
ham_dict = count_words(train_data, 'ham')

# Create a DataFrame from spam_dict and ham_dict
spam_words = pd.DataFrame(list(spam_dict.items()), columns = ['Word', '#Spam'])
ham_words = pd.DataFrame(list(ham_dict.items()), columns = ['Word', '#Ham'])

# Merge the two DataFrames on the 'Word' column
word_freq = pd.merge(spam_words, ham_words, on = 'Word', how = 'outer')

# Fill NaN values with 0
word_freq = word_freq.fillna(0)

# Reset the index
word_freq = word_freq.reset_index(drop = True)

# Removing Decimal Points
word_freq['#Spam'] = word_freq['#Spam'].astype(int)
word_freq['#Ham'] = word_freq['#Ham'].astype(int)
```

<h3>6.  Calculate $P\left(E\middle| S\right)$ and $P\left(E|\lnot S\right)$</h3>
<p>Next create a new DataFrame named word_prob that gives the probability of each word being found in a spam and ham message.</p>
<p>To calculate the probability of a word being spam you divide the number of times the word was found in spam by the total number of spam messages, likewise to calculate the probability of each word being found in a ham message you divide the number of times the word was found in a ham message by the total number of ham messages.</p>
<p>If a word was not found in ham or spam it will cause problems later because the probability calculated will be zero. Therefore, use a pseudocount k and estimate the probability of seeing the word. This is known as smoothing and results in the following formula when k = 0.5, for example.</p>
<p>$P\left(E\middle| S\right)$ = (number of spams containing the word + k) / (total number of spam messages + 2 * k).</p>
<p>Likewise, for $P\left(E|\lnot S\right)$.</p>

```python
total_spam_messages = len(train_data[train_data["Category"] == "spam"])
total_ham_messages = len(train_data[train_data["Category"] == "ham"])
k = 0.5 # Pseudocount 

word_prob = pd.DataFrame(columns=["Word", "P(E|S)", "P(E|¬S)"])

for word in word_freq["Word"]:
    spam_count = word_freq.loc[word_freq['Word'] == word, "#Spam"].iloc[0]
    ham_count = word_freq.loc[word_freq['Word'] == word, "#Ham"].iloc[0]
    
    p_word_given_spam = (spam_count + k) / (total_spam_messages + 2 * k)
    p_word_given_ham = (ham_count + k) / (total_ham_messages + 2 * k)
    
    word_prob = pd.concat([word_prob, pd.DataFrame({"Word": [word], "P(E|S)": p_word_given_spam, "P(E|¬S)": p_word_given_ham})], ignore_index = True)

word_prob.head()
```

<h3>7. Checking the 'spamliness' of a single word</h3>
<p>Now that we have trained the model, we will test the model.  Before we use the test_data, first let’s check how the model calculates the spamliness of a single word.  This is where we use the Bayes Theorem formula.  We have already calculated $P\left(E\middle| S\right)$ and $P\left(E|\lnot S\right)$, so we can just extract these values from the word_prob DataFrame.</p>
<p>We need to decide on the prior values $P\left(S\right)$ and $P\left(\lnot S\right)$, this is where you can experiment and tweak the model, in this example the prior value for spam was set to $0.4$ and the prior value for not spam or ham was set to $0.6$.</p>
<h3>
$P\left(S\middle|\ E\right)=\frac{P\left(E\middle|\ S\right)P\left(S\right)}{P\left(E\middle|\ S\right)P\left(S\right)+P\left(E|\lnot S\right)P\left(\lnot S\right)}$
</h3>
<pre>
Output
Word = ['free']
P(E|S) = [0.29108392]
P(E|¬S) = [0.01365141]
P(S|E) = [0.93427577]
P(¬S|E) = [0.06572423]
</pre>


```python
word = 'free'
p_spam = 0.5 # Prior Value
p_not_spam = 0.5 # Prior Value
p_word_given_spam = word_prob.loc[word_prob['Word'] == word, 'P(E|S)'].iloc[0]
p_word_given_ham = word_prob.loc[word_prob['Word'] == word, 'P(E|¬S)'].iloc[0]

p_spam_given_word = (p_word_given_spam * p_spam) / (p_word_given_spam * p_spam + p_word_given_ham * p_not_spam) # Using formula
p_not_spam_given_word = (p_word_given_ham * p_not_spam) / (p_word_given_spam * p_spam + p_word_given_ham * p_not_spam) # Using formula

# Print the results of the 'spamliness' of a single word
print("Word = {word} \nP(E|S) = {p_word_given_spam} \nP(E|¬S) = {p_word_given_ham} \nP(S|E) = {p_spam_given_word} \nP(¬S|E) = {p_not_spam_given_word}".format(word = word, p_word_given_spam = p_word_given_spam, p_word_given_ham = p_word_given_ham, p_spam_given_word = p_spam_given_word, p_not_spam_given_word = p_not_spam_given_word))
```

<h3>8. Checking the 'spamliness' of several words</h3>
<p>To check the spamliness of several words contained in a message we multiply the probabilities.  The model assumes the words appear as independent events hence the naïve Bayes.  In reality of course, words are not independent events, but the model still performs well.  So we use the assumption that the words appear independently, and hence we multiply probabilities, so
$P(S\,|\, x_1,\dots,x_n)\approx \frac{P(S)\underset{i=1}{\overset{n}{\prod}}P(x_i | S)}{P(S)\underset{i=1}{\overset{n}{\prod}}P(x_i | S)+P(\neg S)\underset{i=1}{\overset{n}{\prod}}P(x_i | \neg S)}$

Calculate the probability for each word in a message being spam, you might want to store the calculations in a list named prob_spam.  Likewise create a list for each word not being spam.
Then multiply the probabilities and compare the results.  If the result of multiplying the probabilities for spam is greater than the result of multiplying the probabilities for not spam, then you assume the message as spam.
</p>
<p>If you have a word in your message that is not in the word_prob DataFrame then you can't get the probability.  Skip any words in the message that are not in the word_prob DataFrame.</p>

```python
# Function to calculate the 'spamliness' of a single word
def calc_spamliness_prob(word):
    p_word_given_spam = word_prob.loc[word_prob['Word'] == word, 'P(E|S)'].iloc[0]
    p_word_given_ham = word_prob.loc[word_prob['Word'] == word, 'P(E|¬S)'].iloc[0]
    
    p_spam_given_word = (p_word_given_spam * p_spam) / (p_word_given_spam * p_spam + p_word_given_ham * p_not_spam) # Using formula
    p_not_spam_given_word = (p_word_given_ham * p_not_spam) / (p_word_given_spam * p_spam + p_word_given_ham * p_not_spam) # Using formula
    
    return p_spam_given_word, p_not_spam_given_word
```

```python
from functools import reduce

message = train_data["Message"][0]
message_type = train_data["Category"][0] # To check whether the model was correct

# Function to calculate the type of the message, either spam or ham
def determine_message_type(message):
    prob_spam = [p_spam]
    prob_ham = [p_not_spam]
    
    for word in message.split():
        # Skip words not in word_prob
        if word not in word_prob["Word"].values:
            continue
            
        p_spam_given_word, p_not_spam_given_word = calc_spamliness_prob(word)
        prob_spam.append(p_spam_given_word)
        prob_ham.append(p_not_spam_given_word)
    
    p_spam_message_denominator = reduce(lambda x,y: x*y, prob_spam) # Multiply all numbers in prob_spam
    p_ham_message_denominator = reduce(lambda x,y: x*y, prob_ham) # Multiply all numbers in prob_ham
    
    p_spam_message = p_spam_message_denominator / (p_spam_message_denominator + p_ham_message_denominator) # Using Formula (the spam message denominator sum is also used as the numerator)
    p_ham_message = p_ham_message_denominator / (p_spam_message_denominator + p_ham_message_denominator) # Using Formula (the ham message denominator sum is also used as the numerator)
    
    print("Probability of spam message: {p_spam_message} \nProbability of ham message: {p_ham_message}".format(p_spam_message = p_spam_message, p_ham_message = p_ham_message))
    
    if (p_spam_message > p_ham_message): return "spam"
    elif (p_spam_message < p_ham_message): return "ham"
    else: return "undetermined"
        
print("Message: {message} \nActual message type: {message_type} \nModel predicted message type: {model_type}".format(message = message, message_type = message_type, model_type = determine_message_type(message))) 
```

<h3>9. Avoiding floating point underflow</h3>
<p>Our aim is to compare two probabilities $P(S|x_1,\dots,x_n)$ with $P(\neg S|x_1,\dots,x_n),$ according to our model introduced in Section 8, both probabilities share a common denominator which does not affect comparison. Hence we will calculate numerators only, which are proportional to $P(S|x_1,\dots,x_n)$ and $P(\neg S|x_1,\dots,x_n).$
</p>

<p>Multiplying a set of small probabilities could result in a floating-point error.  This is where the product becomes too small to be represented correctly.  To avoid this we can take the logarithm of the probabilities and add them.  

To avoid multiplication of small numbers, we use the following property of $\log(x):$</p>
$$
\log(a\cdot b)=\log(a)+\log(b)
$$
<p>i.e. the log of the product is equal to the sum of logs (so instead of multiplying small numbers we will add them):</p>
$$
P(S|x_1,x_2,\dots,x_n)\propto P(S)\cdot P(x_1|S)\cdot \dots \cdot P(x_n|S)$$
<p>becomes</p>
$$\log(P(S|x_1,x_2,\dots,x_n))\propto \log\left(P(S)\cdot P(x_1|S)\cdot \dots  P(x_n|S)\right)=$$ $$
\log(P(S))+\log(P(x_1|S))+\dots+\log(P(x_n|S))
$$
<p>So, to check spam or ham we just compare:</p>
$$
\log(P(S))+\log(P(x_1|S))+\dots+\log(P(x_n|S))
$$
<p>and </p>
$$
\log(P(\neg S))+\log(P(x_1|\neg S))+\dots+\log(P(x_n|\neg S))
$$


Change the equation so that logs are used.
</p>

```python
import numpy as np

message = train_data["Message"][0]
message_type = train_data["Category"][0] # To check whether the model was correct

# Changing equation to logs 
# Function to calculate the type of the message, either spam or ham
def determine_message_type(message):
    prob_spam = [np.log(p_spam)]
    prob_ham = [np.log(p_not_spam)]
    
    for word in message.split():
        # Skip words not in word_prob
        if word not in word_prob["Word"].values:
            continue
            
        p_word_given_spam_log = np.log(word_prob.loc[word_prob['Word'] == word, 'P(E|S)'].iloc[0])
        p_word_given_ham_log = np.log(word_prob.loc[word_prob['Word'] == word, 'P(E|¬S)'].iloc[0])
        
        prob_spam.append(p_word_given_spam_log)
        prob_ham.append(p_word_given_ham_log)
    
    p_spam_message = reduce(lambda x,y: x+y, prob_spam) # Add all logs in prob_spam
    p_ham_message = reduce(lambda x,y: x+y, prob_ham) # Add all logs in prob_ham

    if (p_spam_message > p_ham_message): return "spam"
    else: return "ham"

print("Message: {message} \nActual message type: {message_type} \nModel predicted message type: {model_type}".format(message = message, message_type = message_type, model_type = determine_message_type(message))) 
```

<h3>10. Testing the Model</h3>
<p>Now that we have tested the model using simple messages.  Let’s test the model using the messages from the test_set.  You should implement counters that displays how your model has performed and calculate the accuracy of the model.</p>
<pre>
match_spam 173
match_ham 843
thought_ham_is_spam 3
thought_spam_is_ham 357
Accuracy: 0.7383720930232558
</pre>


```python
match_spam = 0
match_ham = 0
thought_ham_is_spam = 0
thought_spam_is_ham = 0

for index, row in test_data.iterrows():
    message_type = row["Category"] # To check whether the model was correct
    message = row["Message"]
    model_predicted_type = determine_message_type(message)
    
    if model_predicted_type == message_type:
        if message_type == "spam": match_spam += 1
        else: match_ham += 1
    else:
        if message_type == "spam": thought_spam_is_ham += 1
        else: thought_ham_is_spam += 1

accuracy = (match_spam + match_ham) / (match_spam + match_ham + thought_ham_is_spam + thought_spam_is_ham)
print("match_spam: {match_spam} \nmatch_ham: {match_ham} \nthought_ham_is_spam: {thought_ham_is_spam} \nthought_spam_is_ham: {thought_spam_is_ham} \nAccuracy: {accuracy}".format(match_spam = match_spam, match_ham = match_ham, thought_ham_is_spam = thought_ham_is_spam, thought_spam_is_ham = thought_spam_is_ham, accuracy = accuracy))
```

<h3>11. Improvements</h3>

One way to improve the accuracy of the model is to take into account that the probability of words in a given message is not independant. For example, if the word "free" was found in a message the probaility of the word "phone" would be higher than the probability of the word "yellow". Therefore, the probaility of the words are not independant.

Another suggestion is to remove all words that cannot be interpreted as spam or ham. For example, the word "to" provides no meaning in whether the message is spam or ham but could affect the calculations when deciding if the message is spam or ham. Although I did remove some of the common words that appear, it would be better for this to happen automatically with a function for future cases with bigger datasets.

Finally, incorporating more data into the model can also improve its performance. A larger and more diverse dataset can provide more insights into the patterns and characteristics of spam and ham messages. This can lead to a more accurate classification and improved generalization performance.
